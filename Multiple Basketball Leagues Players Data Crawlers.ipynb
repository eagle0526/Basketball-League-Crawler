```python
import requests
import pandas as pd
import bs4
import xlwings as xw


wb_PL = xw.Book(r"D:\專案新檔案_0816\籃球文章\Pleague\Pleague data.xlsx")
# sheet_PL = wb_PL.sheets.add("regular data")

#------------------------------------------------------------例行賽函數-----------------------------------------------------

def PL_regular_crawler(season):
    
    sheet_PL = wb_PL.sheets("regular data")  #新增或讀取sheet表
    res = requests.get(f"https://pleagueofficial.com/stat-player/{season}#record")
    html_PL = bs4.BeautifulSoup(res.text, "html.parser")
    table_PL = html_PL.findAll("table")[0]
    

    #取title列
    table_title_PL = table_PL.findAll("thead")[0].findAll("tr")[0].findAll("th")

    #取球員數據列
    # td_row = table_PL.findAll("tbody")[0].findAll("tr")
    table_data_PL = table_PL.findAll("tbody")[0].findAll("tr")


    #----------title設定 - 先設一個空list，再用append把算出來的加進list裡面

    title_name_PL = [ table_title_PL[i].text for i in range(0, len(table_title_PL)) ]
    title_name_PL.insert(23, "賽季")
    title_name_PL.insert(24, "類型")

    #----------title設定結束


    #----------球員數據設定開始
    for i in range (0,len(table_data_PL)):


        player_Name = table_data_PL[i].findAll("th")[0].text #球員姓名
        player_Number = table_data_PL[i].findAll("td")[0].text #球員號碼
        player_Team = table_data_PL[i].findAll("td")[1].text #球員隊伍
        player_Game = table_data_PL[i].findAll("td")[2].text #出賽次數
        player_Time = table_data_PL[i].findAll("td")[3].text #出賽時間

        player_2FG_Goal = table_data_PL[i].findAll("td")[4].text #兩分進球次數
        player_2FG_Times = table_data_PL[i].findAll("td")[5].text #兩分出手次數
        player_2FG_Rate = table_data_PL[i].findAll("td")[6].text #兩分命中率

        player_3FG_Goal = table_data_PL[i].findAll("td")[7].text #三分進球次數
        player_3FG_Times = table_data_PL[i].findAll("td")[8].text #三分出手次數
        player_3FG_Rate = table_data_PL[i].findAll("td")[9].text #三分命中率

        player_FT_Goal = table_data_PL[i].findAll("td")[10].text #罰球進球次數
        player_FT_Times = table_data_PL[i].findAll("td")[11].text #罰球出手次數
        player_FT_Rate = table_data_PL[i].findAll("td")[12].text #罰球命中率

        player_Score = table_data_PL[i].findAll("td")[13].text #得分
        player_OREB = table_data_PL[i].findAll("td")[14].text #進攻籃板
        player_DREB = table_data_PL[i].findAll("td")[15].text #防守籃板
        player_REB = table_data_PL[i].findAll("td")[16].text #總籃板
        player_AST = table_data_PL[i].findAll("td")[17].text #助攻
        player_STL = table_data_PL[i].findAll("td")[18].text #抄截
        player_BLK = table_data_PL[i].findAll("td")[19].text #阻攻
        player_TO = table_data_PL[i].findAll("td")[20].text #失誤
        player_PF = table_data_PL[i].findAll("td")[21].text #犯規
        Season = (f"{season}")
        Type = "例行賽"


        #把上面那些數據塞進list裡面

        FullPlayer_PL = [ 
            player_Name,    
            player_Number,  
            player_Team,    
            player_Game,    
            player_Time,     
            player_2FG_Goal,     
            player_2FG_Times,
            player_2FG_Rate,
            player_3FG_Goal,     
            player_3FG_Times,
            player_3FG_Rate,
            player_FT_Goal,      
            player_FT_Times, 
            player_FT_Rate, 
            player_Score,   
            player_OREB,    
            player_DREB,    
            player_REB,     
            player_AST,     
            player_STL,     
            player_BLK,     
            player_TO,      
            player_PF,
            Season,
            Type
        ]

#         print(FullPlayer_PL)

    #----------球員數據設定結束


        #第一列數據 - TitleName
        sheet_PL.range("A1").value = title_name_PL

        #第二列數據 - 墊檔用 - **會這樣設定是因為，等等用last_row的方法，沒有第二層程式會跑不出來，只要最後再把第二列刪掉就好
        sheet_PL.range("A2").value = title_name_PL

        #第三列數據 - Full Player
        last_row = sheet_PL.range("A1").end("down").row

        sheet_PL.range(f"A{last_row + 1}").value = FullPlayer_PL[0]
        sheet_PL.range(f"B{last_row + 1}").value = FullPlayer_PL[1]
        sheet_PL.range(f"C{last_row + 1}").value = FullPlayer_PL[2]
        sheet_PL.range(f"D{last_row + 1}").value = FullPlayer_PL[3]
        sheet_PL.range(f"E{last_row + 1}").value = FullPlayer_PL[4]
        sheet_PL.range(f"F{last_row + 1}").value = FullPlayer_PL[5]
        sheet_PL.range(f"G{last_row + 1}").value = FullPlayer_PL[6]
        sheet_PL.range(f"H{last_row + 1}").value = FullPlayer_PL[7]
        sheet_PL.range(f"I{last_row + 1}").value = FullPlayer_PL[8]
        sheet_PL.range(f"J{last_row + 1}").value = FullPlayer_PL[9]
        sheet_PL.range(f"K{last_row + 1}").value = FullPlayer_PL[10]
        sheet_PL.range(f"L{last_row + 1}").value = FullPlayer_PL[11]
        sheet_PL.range(f"M{last_row + 1}").value = FullPlayer_PL[12]
        sheet_PL.range(f"N{last_row + 1}").value = FullPlayer_PL[13]
        sheet_PL.range(f"O{last_row + 1}").value = FullPlayer_PL[14]
        sheet_PL.range(f"P{last_row + 1}").value = FullPlayer_PL[15]
        sheet_PL.range(f"Q{last_row + 1}").value = FullPlayer_PL[16]
        sheet_PL.range(f"R{last_row + 1}").value = FullPlayer_PL[17]
        sheet_PL.range(f"S{last_row + 1}").value = FullPlayer_PL[18]
        sheet_PL.range(f"T{last_row + 1}").value = FullPlayer_PL[19]
        sheet_PL.range(f"U{last_row + 1}").value = FullPlayer_PL[20]
        sheet_PL.range(f"V{last_row + 1}").value = FullPlayer_PL[21]
        sheet_PL.range(f"W{last_row + 1}").value = FullPlayer_PL[22]
        sheet_PL.range(f"X{last_row + 1}").value = FullPlayer_PL[23]
        sheet_PL.range(f"Y{last_row + 1}").value = FullPlayer_PL[24]
        
        
        
#------------------------------------------------------------季後賽函數-----------------------------------------------------
        
def PL_playoff_crawler(season):
    
    sheet_PL = wb_PL.sheets("playoff data")  #新增或讀取sheet表

    res = requests.get(f"https://pleagueofficial.com/stat-player/{season}#record")
    html_PL = bs4.BeautifulSoup(res.text, "html.parser")
    table_PL = html_PL.findAll("table")[0]
    

    #取title列
    table_title_PL = table_PL.findAll("thead")[0].findAll("tr")[0].findAll("th")

    #取球員數據列
    # td_row = table_PL.findAll("tbody")[0].findAll("tr")
    table_data_PL = table_PL.findAll("tbody")[0].findAll("tr")


    #----------title設定 - 先設一個空list，再用append把算出來的加進list裡面

    title_name_PL = [ table_title_PL[i].text for i in range(0, len(table_title_PL)) ]
    title_name_PL.insert(23, "賽季")
    title_name_PL.insert(24, "類型")

    #----------title設定結束


    #----------球員數據設定開始
    for i in range (0,len(table_data_PL)):


        player_Name = table_data_PL[i].findAll("th")[0].text #球員姓名
        player_Number = table_data_PL[i].findAll("td")[0].text #球員號碼
        player_Team = table_data_PL[i].findAll("td")[1].text #球員隊伍
        player_Game = table_data_PL[i].findAll("td")[2].text #出賽次數
        player_Time = table_data_PL[i].findAll("td")[3].text #出賽時間

        player_2FG_Goal = table_data_PL[i].findAll("td")[4].text #兩分進球次數
        player_2FG_Times = table_data_PL[i].findAll("td")[5].text #兩分出手次數
        player_2FG_Rate = table_data_PL[i].findAll("td")[6].text #兩分命中率

        player_3FG_Goal = table_data_PL[i].findAll("td")[7].text #三分進球次數
        player_3FG_Times = table_data_PL[i].findAll("td")[8].text #三分出手次數
        player_3FG_Rate = table_data_PL[i].findAll("td")[9].text #三分命中率

        player_FT_Goal = table_data_PL[i].findAll("td")[10].text #罰球進球次數
        player_FT_Times = table_data_PL[i].findAll("td")[11].text #罰球出手次數
        player_FT_Rate = table_data_PL[i].findAll("td")[12].text #罰球命中率

        player_Score = table_data_PL[i].findAll("td")[13].text #得分
        player_OREB = table_data_PL[i].findAll("td")[14].text #進攻籃板
        player_DREB = table_data_PL[i].findAll("td")[15].text #防守籃板
        player_REB = table_data_PL[i].findAll("td")[16].text #總籃板
        player_AST = table_data_PL[i].findAll("td")[17].text #助攻
        player_STL = table_data_PL[i].findAll("td")[18].text #抄截
        player_BLK = table_data_PL[i].findAll("td")[19].text #阻攻
        player_TO = table_data_PL[i].findAll("td")[20].text #失誤
        player_PF = table_data_PL[i].findAll("td")[21].text #犯規
        Season = (f"{season}")
        Type = "季後賽"


        #把上面那些數據塞進list裡面

        FullPlayer_PL = [ 
            player_Name,    
            player_Number,  
            player_Team,    
            player_Game,    
            player_Time,     
            player_2FG_Goal,     
            player_2FG_Times,
            player_2FG_Rate,
            player_3FG_Goal,     
            player_3FG_Times,
            player_3FG_Rate,
            player_FT_Goal,      
            player_FT_Times, 
            player_FT_Rate, 
            player_Score,   
            player_OREB,    
            player_DREB,    
            player_REB,     
            player_AST,     
            player_STL,     
            player_BLK,     
            player_TO,      
            player_PF,
            Season,
            Type
        ]

#         print(FullPlayer_PL)

    #----------球員數據設定結束


        #第一列數據 - TitleName
        sheet_PL.range("A1").value = title_name_PL

        #第二列數據 - 墊檔用 - **會這樣設定是因為，等等用last_row的方法，沒有第二層程式會跑不出來，只要最後再把第二列刪掉就好
        sheet_PL.range("A2").value = title_name_PL

        #第三列數據 - Full Player
        last_row = sheet_PL.range("A1").end("down").row

        sheet_PL.range(f"A{last_row + 1}").value = FullPlayer_PL[0]
        sheet_PL.range(f"B{last_row + 1}").value = FullPlayer_PL[1]
        sheet_PL.range(f"C{last_row + 1}").value = FullPlayer_PL[2]
        sheet_PL.range(f"D{last_row + 1}").value = FullPlayer_PL[3]
        sheet_PL.range(f"E{last_row + 1}").value = FullPlayer_PL[4]
        sheet_PL.range(f"F{last_row + 1}").value = FullPlayer_PL[5]
        sheet_PL.range(f"G{last_row + 1}").value = FullPlayer_PL[6]
        sheet_PL.range(f"H{last_row + 1}").value = FullPlayer_PL[7]
        sheet_PL.range(f"I{last_row + 1}").value = FullPlayer_PL[8]
        sheet_PL.range(f"J{last_row + 1}").value = FullPlayer_PL[9]
        sheet_PL.range(f"K{last_row + 1}").value = FullPlayer_PL[10]
        sheet_PL.range(f"L{last_row + 1}").value = FullPlayer_PL[11]
        sheet_PL.range(f"M{last_row + 1}").value = FullPlayer_PL[12]
        sheet_PL.range(f"N{last_row + 1}").value = FullPlayer_PL[13]
        sheet_PL.range(f"O{last_row + 1}").value = FullPlayer_PL[14]
        sheet_PL.range(f"P{last_row + 1}").value = FullPlayer_PL[15]
        sheet_PL.range(f"Q{last_row + 1}").value = FullPlayer_PL[16]
        sheet_PL.range(f"R{last_row + 1}").value = FullPlayer_PL[17]
        sheet_PL.range(f"S{last_row + 1}").value = FullPlayer_PL[18]
        sheet_PL.range(f"T{last_row + 1}").value = FullPlayer_PL[19]
        sheet_PL.range(f"U{last_row + 1}").value = FullPlayer_PL[20]
        sheet_PL.range(f"V{last_row + 1}").value = FullPlayer_PL[21]
        sheet_PL.range(f"W{last_row + 1}").value = FullPlayer_PL[22]
        sheet_PL.range(f"X{last_row + 1}").value = FullPlayer_PL[23]
        sheet_PL.range(f"Y{last_row + 1}").value = FullPlayer_PL[24]
        
        
        
#-----------------------------------------------------------冠軍賽函數-----------------------------------------------------
        
def PL_champion_crawler(season):
    
    sheet_PL = wb_PL.sheets("champion data")  #新增或讀取sheet表

    res = requests.get(f"https://pleagueofficial.com/stat-player/{season}#record")
    html_PL = bs4.BeautifulSoup(res.text, "html.parser")
    table_PL = html_PL.findAll("table")[0]
    

    #取title列
    table_title_PL = table_PL.findAll("thead")[0].findAll("tr")[0].findAll("th")

    #取球員數據列
    # td_row = table_PL.findAll("tbody")[0].findAll("tr")
    table_data_PL = table_PL.findAll("tbody")[0].findAll("tr")


    #----------title設定 - 先設一個空list，再用append把算出來的加進list裡面

    title_name_PL = [ table_title_PL[i].text for i in range(0, len(table_title_PL)) ]
    title_name_PL.insert(23, "賽季")
    title_name_PL.insert(24, "類型")

    #----------title設定結束


    #----------球員數據設定開始
    for i in range (0,len(table_data_PL)):


        player_Name = table_data_PL[i].findAll("th")[0].text #球員姓名
        player_Number = table_data_PL[i].findAll("td")[0].text #球員號碼
        player_Team = table_data_PL[i].findAll("td")[1].text #球員隊伍
        player_Game = table_data_PL[i].findAll("td")[2].text #出賽次數
        player_Time = table_data_PL[i].findAll("td")[3].text #出賽時間

        player_2FG_Goal = table_data_PL[i].findAll("td")[4].text #兩分進球次數
        player_2FG_Times = table_data_PL[i].findAll("td")[5].text #兩分出手次數
        player_2FG_Rate = table_data_PL[i].findAll("td")[6].text #兩分命中率

        player_3FG_Goal = table_data_PL[i].findAll("td")[7].text #三分進球次數
        player_3FG_Times = table_data_PL[i].findAll("td")[8].text #三分出手次數
        player_3FG_Rate = table_data_PL[i].findAll("td")[9].text #三分命中率

        player_FT_Goal = table_data_PL[i].findAll("td")[10].text #罰球進球次數
        player_FT_Times = table_data_PL[i].findAll("td")[11].text #罰球出手次數
        player_FT_Rate = table_data_PL[i].findAll("td")[12].text #罰球命中率

        player_Score = table_data_PL[i].findAll("td")[13].text #得分
        player_OREB = table_data_PL[i].findAll("td")[14].text #進攻籃板
        player_DREB = table_data_PL[i].findAll("td")[15].text #防守籃板
        player_REB = table_data_PL[i].findAll("td")[16].text #總籃板
        player_AST = table_data_PL[i].findAll("td")[17].text #助攻
        player_STL = table_data_PL[i].findAll("td")[18].text #抄截
        player_BLK = table_data_PL[i].findAll("td")[19].text #阻攻
        player_TO = table_data_PL[i].findAll("td")[20].text #失誤
        player_PF = table_data_PL[i].findAll("td")[21].text #犯規
        Season = (f"{season}")
        Type = "冠軍賽"


        #把上面那些數據塞進list裡面

        FullPlayer_PL = [ 
            player_Name,    
            player_Number,  
            player_Team,    
            player_Game,    
            player_Time,     
            player_2FG_Goal,     
            player_2FG_Times,
            player_2FG_Rate,
            player_3FG_Goal,     
            player_3FG_Times,
            player_3FG_Rate,
            player_FT_Goal,      
            player_FT_Times, 
            player_FT_Rate, 
            player_Score,   
            player_OREB,    
            player_DREB,    
            player_REB,     
            player_AST,     
            player_STL,     
            player_BLK,     
            player_TO,      
            player_PF,
            Season,
            Type
        ]

#         print(FullPlayer_PL)

    #----------球員數據設定結束


        #第一列數據 - TitleName
        sheet_PL.range("A1").value = title_name_PL

        #第二列數據 - 墊檔用 - **會這樣設定是因為，等等用last_row的方法，沒有第二層程式會跑不出來，只要最後再把第二列刪掉就好
        sheet_PL.range("A2").value = title_name_PL

        #第三列數據 - Full Player
        last_row = sheet_PL.range("A1").end("down").row

        sheet_PL.range(f"A{last_row + 1}").value = FullPlayer_PL[0]
        sheet_PL.range(f"B{last_row + 1}").value = FullPlayer_PL[1]
        sheet_PL.range(f"C{last_row + 1}").value = FullPlayer_PL[2]
        sheet_PL.range(f"D{last_row + 1}").value = FullPlayer_PL[3]
        sheet_PL.range(f"E{last_row + 1}").value = FullPlayer_PL[4]
        sheet_PL.range(f"F{last_row + 1}").value = FullPlayer_PL[5]
        sheet_PL.range(f"G{last_row + 1}").value = FullPlayer_PL[6]
        sheet_PL.range(f"H{last_row + 1}").value = FullPlayer_PL[7]
        sheet_PL.range(f"I{last_row + 1}").value = FullPlayer_PL[8]
        sheet_PL.range(f"J{last_row + 1}").value = FullPlayer_PL[9]
        sheet_PL.range(f"K{last_row + 1}").value = FullPlayer_PL[10]
        sheet_PL.range(f"L{last_row + 1}").value = FullPlayer_PL[11]
        sheet_PL.range(f"M{last_row + 1}").value = FullPlayer_PL[12]
        sheet_PL.range(f"N{last_row + 1}").value = FullPlayer_PL[13]
        sheet_PL.range(f"O{last_row + 1}").value = FullPlayer_PL[14]
        sheet_PL.range(f"P{last_row + 1}").value = FullPlayer_PL[15]
        sheet_PL.range(f"Q{last_row + 1}").value = FullPlayer_PL[16]
        sheet_PL.range(f"R{last_row + 1}").value = FullPlayer_PL[17]
        sheet_PL.range(f"S{last_row + 1}").value = FullPlayer_PL[18]
        sheet_PL.range(f"T{last_row + 1}").value = FullPlayer_PL[19]
        sheet_PL.range(f"U{last_row + 1}").value = FullPlayer_PL[20]
        sheet_PL.range(f"V{last_row + 1}").value = FullPlayer_PL[21]
        sheet_PL.range(f"W{last_row + 1}").value = FullPlayer_PL[22]
        sheet_PL.range(f"X{last_row + 1}").value = FullPlayer_PL[23]
        sheet_PL.range(f"Y{last_row + 1}").value = FullPlayer_PL[24]



#------------------------------------------------------------引用例行賽函數-----------------------------------------------------
        
#PL所有賽季
PL_Full_Season = ["2020-21/2",   #第一季 
                  "2021-22/6",   #第二季
                 ]

#把每個數字都拉出來後，丟到參數裡面
for i in PL_Full_Season:
    PL_regular_crawler(i)       #啟動函數
    


#------------------------------------------------------------引用季後賽函數-----------------------------------------------------

#PL所有賽季
PL_Full_Season = ["2020-21/3",   #第一季
                  "2021-22/7"    #第二季
                 ]

#把每個數字都拉出來後，丟到參數裡面
# for i in PL_Full_Season:
#     PL_playoff_crawler(i)       #啟動函數
    
    
#------------------------------------------------------------引用冠軍賽函數-----------------------------------------------------

#PL所有賽季
PL_Full_Season = ["2020-21/4",   #第一季
                  "2021-22/7"    #第二季
                 ]

#把每個數字都拉出來後，丟到參數裡面
# for i in PL_Full_Season:
#     PL_champion_crawler(i)    #啟動函數
        
        
# #刪掉第二列
# sheet_PL["2:2"].delete()



```

```python
import pandas as pd
import xlwings as xw


wb_cba = xw.Book(r"D:\專案新檔案_0816\籃球文章\中國CBA\CBA.xlsx")

def cba_crawler(season):
    
    sheet_cba = wb_cba.sheets.add(f"{season}")
    
    #用DataFrame直接抓數據
    df_cba = pd.read_html(

        f"http://cba.sports.sina.com.cn/cba/stats/playerstats/?qleagueid={season}&qround=0", #資料來源
        encoding='gbk', #中文編碼
    )

    df_cba = df_cba[0]
    df_one = df_cba["罚球"].str.split("-").str.get(0)
    df_two = df_cba["罚球"].str.split("-").str.get(1)
    df_cba["FTM"] = df_one
    df_cba["FTA"] = df_two
    df_cba["FTM"] = df_cba["FTM"].astype(float)
    df_cba["FTA"] = df_cba["FTA"].astype(float)
    df_cba["FT%"] = round((df_cba["FTM"] / df_cba["FTA"]), 4)
    
    df_cba["season"] = (f"{season}")    
    df_cba["League"] = "CBA"
    
    sheet_cba.range("A1").value = df_cba




CBA_season = {
    "16-17" : 180,
    "17-18" : 189,
    "18-19" : 198,
    "19-20" : 205,
    "20-21" : 206,
    "21-22" : 210
}

value = CBA_season.values()
value

for season in value:
    cba_crawler(season)
```
